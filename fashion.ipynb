```
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
```

```
(trainx,trainy),(testx,testy)=keras.datasets.fashion_mnist.load_data()
trainx=trainx/255
testx=testx/255
print(trainx.shape)
```

```
model = keras.Sequential([
              keras.layers.Flatten(input_shape=(28,28)),
              keras.layers.Dense(128,activation="relu"),
              keras.layers.Dense(10,activation="softmax")
              ])

model.compile(optimizer="adam", loss="sparse_categorical_crossentropy", metrics=["accuracy"])

model.fit(trainx,trainy,epochs=200)
```

```
from google.colab import drive
drive.mount('/content/gdrive')
dir='/content/gdrive/My Drive/Colab Notebooks/im('
from PIL import Image
customx=[]
for i in range(10):
  d2=dir+str(i)+').jpg'
  image = Image.open(d2).convert('L')
  data = np.asarray(image)
  data=data/255
  data=1-data
  customx.append(data)
customx=np.array(customx)
print(customx.shape)
```

```
prediction=model.predict(testx)
j=0
k=0
for i in prediction:
  if(np.argmax(i)==testy[j]):
    k+=1
  j+=1
print("Accuracy on test data = "+str(k/j*100)+" %")
```

```
model.evaluate(testx)
```
